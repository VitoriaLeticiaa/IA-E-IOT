```python
import pandas as pd
import seaborn as se
```

```python
agendamento = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/agendamento3.csv")
```

```python
agendamento.head()
```

```python
import matplotlib.pyplot as plt

agendamento['Curso'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Agendamentos por Curso')
plt.xlabel('Curso')
plt.ylabel('Quantidade')
plt.show()
```

```python
agendamento['Módulo'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title('Agendamentos por Módulo')
plt.xlabel('Módulo')
plt.ylabel('Quantidade')
plt.show()
```
